```python
from datetime import datetime
import pandas as pd
```

```python
df = pd.read_csv('id-cleaned-tweet-data.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
df['year'] = df.Date.apply(lambda x: x.strftime('%Y'))
df['month'] = df.Date.apply(lambda x: x.strftime('%m'))
df.head()
```

```python
df['month'].value_counts()
```

```python
df = pd.DataFrame(df)
df.to_csv('id-cleaned-tweet-data.csv')
```
