```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('test-999.csv', encoding = 'latin-1')
df.head()
```

```python
df.info()
```

```python
counter = Counter(df['common_tweet'].values)
print(counter.most_common(10))
```

```python
cnt = Counter()

for q in df['common_tweet'][df['sentiment_score']==-1].values:
    for o in str(q).split():
        cnt[o]+=1
cnt.most_common(10)
```

```python
for y in df['common_tweet'][df['sentiment_score']==0].values:
    for u in str(y).split():
        cnt[u]+=1
cnt.most_common(10)
```

```python
for i in df['common_tweet'][df['sentiment_score']==1].values:
    for w in str(i).split():
        cnt[w]+=1
cnt.most_common(10)
```
